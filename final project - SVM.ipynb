### splitting data into dependent and independent variables

```python
X = df.drop('DIAG', axis=1).copy()
X.head()
```

```python
y = df['DIAG'].copy()
y.head()
```

### scaling the data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
X_train_scaled = scale(X_train)
X_test_scaled = scale(X_test)
```

```python
X_train.head()
```

```python
X_train_scaled
```

### begin SVM 

```python
start_svm = SVC(random_state=2)
start_svm.fit(X_train_scaled, y_train)
```

```python
plot_confusion_matrix(start_svm,
                     X_test_scaled,
                     y_test,
                     values_format='d',
                     display_labels=['No CC diagnosis', 'CC diagnosis'])
```

```python
param_grid = [{'C': [0.1, 1, 10, 100, 1000],
              'gamma': ['scale', 10, 1, 0.1, 0.01, 0.001, 0.0001,'auto'], 
              'kernel': ['poly', 'rbf', 'sigmoid']}]
```

```python
optimal_params = GridSearchCV(SVC(),
                             param_grid,
                             cv=5,
                             scoring='accuracy',
                             verbose=0)
```

```python
optimal_params.fit(X_train_scaled, y_train)
print(optimal_params.best_params_)
```

```python
opt_SVM = SVC(random_state=2, C=1, gamma=0.01, kernel='rbf')
```

```python
opt_SVM.fit(X_train_scaled, y_train)
```

```python
plot_confusion_matrix(opt_SVM,
                     X_test_scaled,
                     y_test,
                     values_format='d',
                     display_labels=['No CC diagnosis', 'CC diagnosis'])
```

```python
pca = PCA()
X_train_PCA = pca.fit_transform(X_train_scaled)

per_var = np.round(pca.explained_variance_ratio_*100, decimals=1)
labels = [str(x) for x in range(1, len(per_var) + 1)]
```

```python
plt.bar(x=range(1, len(per_var) +1), height=per_var)
plt.tick_params(
    axis='x',
    which='both',
    bottom=False,
    top=False,
    labelbottom=False)
plt.ylabel('% of explained variance')
plt.xlabel('Principal components')
plt.show()
```

```python
PC_train_feat_1 = X_train_PCA[:,0]   #x-coordinates after PCA
PC_train_feat_2 = X_train_PCA[:,1]   #y-coordinates after PCA
```

```python
PCA_train_scaled = scale(np.column_stack((PC_train_feat_1, PC_train_feat_2)))
```

```python
optimal_params = GridSearchCV(SVC(),
                             param_grid,
                             cv=5,
                             scoring='accuracy',
                             verbose=2)
```

```python
PCA_test_scaled = pca.transform(X_train_scaled)

PC_test_feat_1 = PCA_test_scaled[:,0]   #x-coordinates after PCA
PC_test_feat_2 = PCA_test_scaled[:,1]   #y-coordinates after PCA
```

Here, we make a matrix so we can show the 'decision regions'. We make this matrix slightly larger than the transformed PCA points so they'll fit inside the region. 

```python
x_min = PC_test_feat_1.min() - 1
x_max = PC_test_feat_1.max() + 1

y_min = PC_test_feat_2.min() - 1
y_max = PC_test_feat_2.max() + 1
```

```python
xx, yy = np.meshgrid(np.arange(start=x_min, stop=x_max, step=0.1),
                    np.arange(start=y_min, stop=y_max, step=0.1))
```

```python
z = PCA_svm.predict(np.column_stack((xx.ravel(), yy.ravel())))
```

```python
z = z.reshape(xx.shape)
```

```python
fig, ax = plt.subplots(figsize=(6,6))

ax.contourf(xx, yy, z, alpha=0.1)
cmap = colors.ListedColormap(['#0000FF', '#00FF00'])

scatter = ax.scatter(PC_test_feat_1, PC_test_feat_2,
                    c = y_train, 
                    cmap = cmap,
                    s = 100,
                    edgecolors = 'k',
                    alpha = 0.7)

legend = ax.legend(scatter.legend_elements()[0],
                  scatter.legend_elements()[1],
                  loc='upper right')
legend.get_texts()[0].set_text('No CC diagnosis')
legend.get_texts()[1].set_text('CC diagnosis')

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')

ax.set_title('SVM decision surface using PCA features')

plt.show()
```
